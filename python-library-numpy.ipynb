```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
sn.distplot([0,1,2,3,4,5])
plt.show()
```

```python
sn.distplot([0,1,2,3,4,5],hist=False)
plt.show()
```

# Normal Distribution

```python
from numpy import random
x=random.normal(size=(2,3))
print(x)
```

```python
y=sn.distplot(random.normal(size=5000),hist=False)
```

# Binomial Distribution

```python
z=random.binomial(10,0.5,10)
print(z)
```

```python
zz=sn.distplot(random.binomial(10,0.5,10000),hist=True,kde=False)
```

# Poisson Distribution

```python
a=random.poisson(2,10)
print(a)
```

```python
aa=sn.distplot(random.poisson(2,10000),kde=False)
plt.show()
```

# Uniform Distribution

```python
w=random.uniform(size=(2,3))
print(w)
```

```python
aa=sn.distplot(random.uniform(size=1000),hist=False)
```

# Multinomial Distribution

```python
x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)
```

# Exponential Distribution

```python
x=random.exponential(scale=2,size=(2,3))
print(x)
```

```python
asa=sn.distplot(random.exponential(size=1000),hist=False)
```

# Chi Square Distribution

```python
x=random.chisquare(df=2,size=(2,3))
print(x)
```

```python
aso=sn.distplot(random.chisquare(df=1,size=1000),hist=False)
```
